## Model Building 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

## -----> OLS Regression

```python
from statsmodels.regression.linear_model import OLS
import statsmodels.regression.linear_model as smf
```

```python
regression=smf.OLS(endog=y_train,exog=X_train).fit()
regression.summary()
```

#### Evaluation for OLS

```python
y_test_pred=regression.predict(X_test)
y_train_pred=regression.predict(X_train)
```

```python
from sklearn.metrics import r2_score
print('Test Data Accuracy using OLS: ',r2_score(y_test,y_test_pred))
print('Train Data Accuracy using OLS: ',r2_score(y_train,y_train_pred))
```

## -----> Regularisation Techniques

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
rid=Ridge(alpha=0.3)
las=Lasso(alpha=0.3)
ela=ElasticNet(alpha=0.3,l1_ratio=0.3)
```

```python
rid.fit(X_train,y_train)
las.fit(X_train,y_train)
ela.fit(X_train,y_train)
```

#### Evaluation for Ridge, Lasso and ElasticNet

```python
y_test_pred=rid.predict(X_test)
y_train_pred=rid.predict(X_train)
print('Test Data Accuracy using Ridge : ',r2_score(y_test,y_test_pred))
print('Train Data Accuracy using Ridge: ',r2_score(y_train,y_train_pred))
y_test_pred=las.predict(X_test)
y_train_pred=las.predict(X_train)
print('\nTest Data Accuracy using Lasso: ',r2_score(y_test,y_test_pred))
print('Train Data Accuracy using Lasso: ',r2_score(y_train,y_train_pred))
y_test_pred=ela.predict(X_test)
y_train_pred=ela.predict(X_train)
print('\nTest Data Accuracy using ElasticNet: ',r2_score(y_test,y_test_pred))
print('Train Data Accuracy using ElasticNet: ',r2_score(y_train,y_train_pred))
```

        if not linear then it may be a polynomial Regression Problem

```python
sns.scatterplot(x=y_train,y=x_train_pred)
plt.show()
```

#### 2) No or little multicolinearity: 
        The independent variables should not depend on each other
        check using Variance Inflation Factor (VIF) > 5

#### 3) No Hetroscadicity: 
        The distribution plot should be normaly distributed 

```python
sns.displot((y_train-y_train_pred),bins=50)
plt.show()
```

#### 4) No autocorrelation: 
        Check of Durbin watson constant in OLS
       if 1.5 < DW < 2.5 it is a Regression problem
       else a Time Series problem
            ARIMA , AutoRegressor can be used for Time Series

#### 5) No Endoginity:
